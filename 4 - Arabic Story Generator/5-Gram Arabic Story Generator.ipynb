# 5-Gram Arabic Story Generator

```python
from nltk.util import ngrams
from google.colab import drive

def load_splitted_corpus(file_name):
    base_url = "url or Drive Path/"
    text = open(base_url+file_name, encoding='utf-8').read().split()
    return text

#drive.mount('/content/drive')
```

```python
file_path = 'story_corpus_qisasse_com.txt'
tokens = load_splitted_corpus(file_path)

output = list(ngrams(tokens, 5))
print (output)
```
