# Simple Arabic Non-Goal Driven Chatbot
## Using ReGex

Import re

```
import re
```

declare pairs of inputs and outputs, with text normalization

```
pairs = [
        [r'(.*)حزين(.*)', 'أنا آسفٌ لسماع مثل هذا؛ لماذا؟'],
        [r'(.*)سعيد(.*)', 'هذا جيد. وأنا سعيدٌ لسعادتك'],
        [r'(.*)اسمي(.*)', 'سررت بمعرفتك'],
        [r'(.*)حالك(.*)', 'بخير وبدون Bugs :)'],
        [r'(.*)هل أنت بخير(.*)', 'أجل. بخير'],
        [r'(.*)مرحبا(.*)', 'أهلاً بك'],
        [r'(.*)عمرك(.*)', 'لا أدري؛ حسبما يتنهي تنفيذ البرنامج'],
        [r'(.*)اسمك(.*)', 'اسمي روبوت المحادثة الآلية'],
        [r'(.*)مكانك(.*)', 'على قرص D :)'],
        [r'(.*)هواياتك(.*)', 'البرم فقط'],
        [r'(.*)لغتك(.*)', 'العربية'],
        [r'(.*)دراس(.*)', 'NLP أولاً وأخيراً'],
        [r'(.*)عملك(.*)', 'أبرم معك بس']
        ]
```

chatting loop

```
while True:
    user_input = input ('')

    if user_input != '0':
        is_found = False
        for p in pairs:
            match_object = re.search(p[0], user_input, flags=re.IGNORECASE)
            if match_object:
                print ('> ' + p[1])
                is_found = True
                break
        
        if not is_found:
            print ('> إنني أحاول فهم ما تقول. لا عليك')
            
    else:
        print ('> أراك قريباً. بالسلامة')
        break
    
    print ()
```
